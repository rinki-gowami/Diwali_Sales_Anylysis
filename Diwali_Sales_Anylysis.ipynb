```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
ax=sns.countplot(x='Gender',data=df,color='#ff4000' )
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()    
```

```python
sales_gender=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sales_gender
```

```python
sns.barplot(x='Gender',y='Amount',data=sales_gender,color='#E5FB95')
plt.show()
```

```python
df.columns
```

```python
sns.countplot(x='Age Group',data=df)
plt.show()
```

```python
ax=sns.countplot(x='Age Group',hue='Gender',data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#total amount vs age group
```

```python
sales_age=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
sns.barplot(x='Age Group',y='Amount',data=sales_age)
plt.show()
```

```python
df.columns
```

```python
sales_state=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
ax=sns.barplot(x='State',y='Orders',data=sales_state)
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

```python
#Total sales form top 10 states
```

```python
sales_data=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.barplot(x='State',y='Amount',data=sales_data)
plt.show()
```

```python
ax=sns.countplot(x='Marital_Status',data=df)
sns.set(rc={'figure.figsize':(6,4)})
```

```python
sales_data_by_martial_status=df.groupby(['Gender','Marital_Status'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(5,5)})
sns.barplot(x='Marital_Status',y='Amount',data=sales_data_by_martial_status,hue='Gender')
```

```python
sns.set(rc={'figure.figsize':(20,6)})
ax=sns.countplot(x='Occupation',data=df)
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()    
```

```python
occupation_based_data=df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(20,7)})
sns.barplot(x='Occupation',y='Amount',data=occupation_based_data)
plt.show()
```

```python
df.columns
```

```python
ax=sns.countplot(x='Product_Category',data=df)
sns.set(rc={'figure.figsize':(23,6)})
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()    
```

```python
product_sales_data=df.groupby(['Product_Category'],as_index=False).sum().sort_values(by='Amount',ascending=False).head(10)
sns.barplot(x='Product_Category',y='Amount',data=product_sales_data)
sns.set(rc={'figure.figsize':(10,10)})
plt.show()
```

```python
sales_Data=df.groupby(['Product_ID'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.barplot(x='Product_ID',y='Orders',data=sales_Data)
plt.show()
```
